```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
extraversion_selfesteem_data = pd.read_csv('Sample data - Sheet4(1).csv')
age_data = pd.read_csv('Age Data.csv')  
gender_data = pd.read_csv('Gender Data.csv') 

merged_data = pd.concat([extraversion_selfesteem_data, age_data['Age'], gender_data['Gender']], axis=1)

merged_data.info()
merged_data.head()
```

```python
print("Descriptive Statistics for Extraversion, Self-Esteem, and Age:\n")
print(merged_data[['EPQ-R Scores', 'RSE Scores', 'Age']].describe())

extraversion_median = merged_data['EPQ-R Scores'].median()
selfesteem_median = merged_data['RSE Scores'].median()
extraversion_mode = merged_data['EPQ-R Scores'].mode()[0]
selfesteem_mode = merged_data['RSE Scores'].mode()[0]

print(f"\nExtraversion - Median: {extraversion_median}, Mode: {extraversion_mode}")
print(f"Self-Esteem - Median: {selfesteem_median}, Mode: {selfesteem_mode}")
```

```python
correlation = merged_data['EPQ-R Scores'].corr(merged_data['RSE Scores'])
print(f"\nPearson Correlation between Extraversion and Self-Esteem: {correlation}")
```

```python
print("\nGrouped Statistics for Extraversion and Self-Esteem by Gender:")
gender_grouped_stats = merged_data.groupby('Gender')[['EPQ-R Scores', 'RSE Scores']].mean()
print(gender_grouped_stats)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(merged_data['EPQ-R Scores'], merged_data['RSE Scores'], color='blue', label='Data points')
m, b = np.polyfit(merged_data['EPQ-R Scores'], merged_data['RSE Scores'], 1)
plt.plot(merged_data['EPQ-R Scores'], m * merged_data['EPQ-R Scores'] + b, color='red', label='Trend line')
plt.title("Extraversion vs Self-Esteem")
plt.xlabel("Extraversion Score")
plt.ylabel("Self-Esteem Score")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.boxplot(merged_data['EPQ-R Scores'], vert=False)
plt.title('Box Plot for Extraversion')
plt.xlabel("Extraversion Score")

plt.subplot(1, 2, 2)
plt.boxplot(merged_data['RSE Scores'], vert=False)
plt.title('Box Plot for Self-Esteem')
plt.xlabel("Self-Esteem Score")

plt.tight_layout()
plt.show()
```

```python
plt.hist(age_data['Age'], bins = 10, color = 'skyblue', edgecolor = 'black')
plt.title('Age distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.boxplot(age_data['Age'], vert=False, patch_artist=True)
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(merged_data['Age'], merged_data['RSE Scores'], color='purple')
plt.title("Age vs Self-Esteem")
plt.xlabel("Age")
plt.ylabel("Self-Esteem Score")
plt.show()
```
